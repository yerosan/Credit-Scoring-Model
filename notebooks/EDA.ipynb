### Data Insights from Visual EDA

#### 1. Numerical Features Distribution
- **Amount** and **Value** show a right-skewed distribution due to high transaction amounts.
- Most transactions are under $1,000, with a few extreme values.

#### 2. Outliers
- Outliers in **Amount** and **Value** (above $1,000,000) affect the dataset's mean and standard deviation.

#### 3. Correlation Analysis
- Strong correlation between **Amount** and **Value** suggests redundancy.
- Low correlation between **FraudResult** and other features indicates weak linear relationships with fraud detection.

#### 4. Categorical Feature Distribution
- **ProductCategory**: Dominated by "Financial Services" and "Airtime".
- **ChannelId**: "ChannelId_3" handles the bulk of transactions.

#### 5. Outliers & Missing Data
- Outliers could skew model predictions (consider trimming/winsorization).
- No missing data in the dataset.
